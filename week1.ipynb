$r_t=ε_t+θ ε_{t-1}$

$p_t=p_{t−1}+r_t$

if $ε_t=0$ for all t except $ε_1=1$

for the case 
$θ=+0.5$, ("momentum") 

$θ=0$ ("i.i.d." or "random walk")

$θ=−0.5$ ("mean reversion") 

Assume μ=0 and $p_0=0$ (think of it as the log price)


```python
def shock_epsilon(theta, steps=10, shock_t=1, shock_ample=1):
    epsilon = zeros(steps)
    epsilon[shock_t] = shock_ample
    r_t = zeros(steps)
    r_t[1:steps] = epsilon[1:steps] + theta * epsilon[0:steps-1]   
    return r_t

def plot_rt(theta):
    rt = shock_epsilon(theta)
    pt = cumsum(rt)
    plot(rt)
    plt.show()
    print("theta={}, returns={}".format(theta, rt[1:6]))
    plot(pt)
    plt.show()
    print("theta={}, p_t={}".format(theta, pt[1:6]))

plot_rt(0.5)
plot_rt(0)      
plot_rt(-0.5)
    
```

If x is normally-distributed (so that ex is lognormal), then

$E(e^x)=e^{\mu + 1/2 σ^2}\\
median(e^x)=e^\mu\\
mode(e^x)=e^{\mu−σ^2}\\
Pr(e^x<e^y)=Pr(x<y)$

```python
mu = 0.06
sigma = 0.20
ar = lambda T:exp(mu*T)-1
av = lambda T:(ar(T)+1)*sqrt(exp(T*0.2**2)-1)
asharpe = lambda T:ar(T) / av(T)

Ex = lambda T:ar(T)+1
median = lambda T:exp(mu*T-1/2*sigma**2*T)
mode = lambda T:exp((mu-3/2*sigma**2)*T)
from scipy import stats
logmean = lambda T:mu*T
lessthanmean = lambda T: stats.norm(loc=mu*T-1/2*sigma**2*T,scale=sqrt(T)*sigma).cdf(logmean(T))
```

```python
def printmy(T):
    print("reture, volatility, sharpe={:.2f} {:.2f} {:.2f}".format(ar(T)*100, av(T)*100, asharpe(T)))
printmy(1)
printmy(10)
printmy(100)
```

```python
T = arange(1,100)
plot(T,ar(T),T,T)
plt.show()
plot(T,av(T),T,sqrt(T))
plt.show()
plot(T,asharpe(T),T,sqrt(T))
plt.show()
plot(T,lessthanmean(T))
```

```python
def domy(T=10):
    t=arange(T)
    if T >= 10:
        plot(t,Ex(t), t, median(t), t, mode(t))
        plt.show()
    print(T,Ex(T), median(T),mode(T),lessthanmean(T)*100)

for i in [1, 5, 10, 50]:
    domy(i)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def myfit(formular='Return ~ Return.shift()', data=a):
    results=smf.ols(formula=formular, data=data).fit()
    res = [results.params[1], 
           results.tvalues[1],
           results.rsquared, 
           100 * results.fittedvalues.mean(),
           100 * results.fittedvalues.std(),
           #100 * sqrt(results.scale),
           ]
    print(' '.join('{:.2f}'.format(i) for i in res))
    #print(results.summary())
    return results
myfit()
myfit('Treturn ~ Treturn.shift()')
myfit('Excess ~ Excess.shift()')
myfit('Excess ~ DP.shift()')
```

```python
tot_ret = ret + 1
tot_tret = tret + 1
from functools import reduce
from operator import mul
a['ret5'] = reduce(mul,[tot_ret.shift(-i) for i in range(1, 6)])
a['tret5'] = reduce(mul,[tot_tret.shift(-i) for i in range(1, 6)])
a['excess5'] = a['ret5'] - a['tret5']
```

```python
res = myfit('excess5 ~ DP')

```

```python
def test_ols_assumptions(res):
    X = res.model.data.orig_exog
    y = res.model.data.orig_endog
    print(res.summary())
    print()
    
    t1 = '''1. Strict exogeneity: E(e|X) = 0
    Use Instrumental variable if violated
    1.1 E(e) = 0
    1.2 the regressors are uncorrelated with the error: E(X'e) = 0
    2 No linear dependence
    '''
    #print('resid-X corr=%s' % res.resid.corr(a['DP']))
    print(t1)
    
    corr_matrix = pd.concat([pd.DataFrame({'error': res.resid}), X.drop('Intercept', axis=1), y], axis=1).corr()
    print('Corr Matrix, should be small except diagonals or dependants:\n', corr_matrix)
    
    scatter(x=res.fittedvalues, y=res.resid)
    plot([min(res.fittedvalues), max(res.fittedvalues)],[0,0])
    title('residual - fit: should not show any structures')
    show()
    fig = sm.graphics.plot_ccpr_grid(res)
    show()
    
    t3 = '''3 Spherical errors: Var[e|X] = \sigma^2 * I
    3.1 Homoscedasticity: E[ e_i^2 | X ] = \sigma^2, same variance in each observation
    If found Heteroscedasticity use robust estimation
    '''
    print(t3)
    print('hetbp(Breush Pagan) Heteroscedasticity test, p-value should be large:')
    print(sm.stats.het_breushpagan(res.resid, res.model.exog))
    
    for k,x in X.drop('Intercept', axis=1).iteritems():
        scatter(x,y)
        title('y - x({}) plot: should not show structure'.format(k))
        show()
        sm.graphics.plot_fit(res, k)
        show()
        
    t32 = '''3.2 Nonautocorrelation:  E[ e_i e_j | X ] = 0
    If found autocorrelation, use GLS for improved efficiency
    '''
    print("test for autocorrelation:")
    print('Breush Godfrey test, p-value should be large', sm.stats.acorr_breush_godfrey(res))
    print('L Jungbox test, p-value should be large', sm.stats.acorr_ljungbox(res.resid, boxpierce=True))
    
    t4 = '''4. Normality: e|X ~ N(0, \sigma^2 I)
    for hypothesis test
    '''
    print(t4)
    fig = sm.qqplot(res.resid, line='r')
    title('qqplot for residuals')
    show()



test_ols_assumptions(res)

```
